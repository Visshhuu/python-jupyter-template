### Exercise 1: Line Plot Customization

```python
import matplotlib.pyplot as plt
import numpy as np

# Data
x = np.arange(0, 11)
y = x ** 2

# Create the plot
plt.plot(x, y, marker='o', linestyle='-', color='b', label='$y = x^2$')

# Customize the plot
plt.title("Square Function Plot")
plt.xlabel("X values")
plt.ylabel("Y values (Square of X)")
plt.grid(True)
plt.legend()

# Show the plot
plt.show()
```
