### Exercise 5: Subplot Layout

```python
import matplotlib.pyplot as plt
import numpy as np

# Data
x = np.arange(1, 6)
y = x * 2
categories = ['A', 'B', 'C', 'D']
values = [5, 7, 3, 9]
labels = ['Python', 'Java', 'C++', 'JavaScript']
sizes = [40, 25, 20, 15]
colors = ['blue', 'red', 'green', 'orange']

# Create subplots
fig, axs = plt.subplots(2, 2, figsize=(10, 10))

# Line plot (Top-Left)
axs[0, 0].plot(x, y, marker='o', linestyle='-', color='b')
axs[0, 0].set_title("Line Chart")
axs[0, 0].set_xlabel("X values")
axs[0, 0].set_ylabel("Y values")
axs[0, 0].grid(True)

# Bar chart (Top-Right)
axs[0, 1].bar(categories, values, color=['red', 'blue', 'green', 'purple'])
axs[0, 1].set_title("Bar Chart")
axs[0, 1].set_xlabel("Categories")
axs[0, 1].set_ylabel("Values")
axs[0, 1].grid(axis='y', linestyle='--', alpha=0.7)

# Scatter plot (Bottom-Left)
axs[1, 0].scatter(x, y, color='blue', marker='o')
axs[1, 0].set_title("Scatter Plot")
axs[1, 0].set_xlabel("X values")
axs[1, 0].set_ylabel("Y values")
axs[1, 0].grid(True)
for i in range(len(x)):
    axs[1, 0].text(x[i], y[i], f'({x[i]}, {y[i]})', fontsize=10, ha='right', va='bottom')

    # Pie chart (Bottom-Right)
    axs[1, 1].pie(sizes, labels=labels, autopct='%1.1f%%', colors=colors, startangle=140, shadow=True)
    axs[1, 1].set_title("Pie Chart")

    # Adjust layout for better spacing
    plt.tight_layout()

    # Show the plot
    plt.show()
    
```
